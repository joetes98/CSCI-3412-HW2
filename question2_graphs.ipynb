```python
import matplotlib.pyplot as plt
import pandas as pd

d = {'Insertion Sort': [0.021, 2.345, 240.403, 1821.183, 8069.402, 39616.518],
         'Merge Sort': [0.003, 0.039, 0.482, 1.525, 3.011, 6.714] }
    
rowLabels = [1000, 10000, 100000, 250000, 500000, 1000000]
df = pd.DataFrame(data = d)
df.index = rowLabels

# insertiondf = df.iloc[:, :1]
# insertion = insertiondf.plot.bar(rot=0)

# mergedf = df.iloc[:, -1]
# merge = mergedf.plot.bar(rot=0)
```

```python
# insertion sort graph

insertiondf = df.iloc[:, :1]
insertion = insertiondf.plot(color ='b', marker ='o', mfc = 'r', mec = 'black')

insertion.set_title('Execution time for insertion sort')
insertion.set_xlabel('Data size')
insertion.set_ylabel('Execution time (seconds)')
```

```python
# merge sort graph

mergedf = df.iloc[:, 1:]
merge = mergedf.plot(color ='b', marker ='o', mfc = 'r', mec = 'black')
merge.set_title('Execution time for merge sort')
merge.set_xlabel("Data Size")
merge.set_ylabel("Execution time (seconds)")
```
